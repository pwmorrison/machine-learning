# Most bagels

```python
df_bagels = df.loc[:, ['tourney_id', 'match_num', 'winner_name', 'loser_name', 'score']]

# The matches that contain a bagel (6-0 or 0-6). I think the first score corresponds to the match winner.
# The set score must appear at the start of the string, or be preceded by a space. This removes the tie-breakers that contain things like '10-6'
df_60 = df_bagels.loc[df['score'].str.contains('(^| )6-0', regex=True)]
df_06 = df_bagels.loc[df['score'].str.contains('(^| )0-6', regex=True)]

# Create a new column that contains the player that got the bagel/s.
df_60['player'] = df_60['winner_name']
df_06['player'] = df_06['loser_name']

# Create a new column that contains the number of bagels.
df_60['num_bagels'] = df_60['score'].str.count('(^| )6-0')
df_06['num_bagels'] = df_06['score'].str.count('(^| )0-6')

# Merge the two dataframes.
df_bagels = pd.concat([df_60, df_06])

# Sum the bagels per person.
df_bagels = df_bagels.groupby('player')['num_bagels'].sum()

df_bagels.iloc[0:5]
df_bagels
```
